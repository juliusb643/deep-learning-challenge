```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
columns_to_drop = ['EIN', 'NAME']
application_df.drop(columns_to_drop, axis=1, inplace=True)
application_df.head()
```

```python
unique_value_counts = application_df.nunique()
print(unique_value_counts)
```

```python
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
threshold = 50
application_types_to_replace = application_type_counts[application_type_counts < threshold].index
application_df['APPLICATION_TYPE'].replace(application_types_to_replace, 'Other', inplace=True)
application_df['APPLICATION_TYPE'].value_counts()
```

```python
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
threshold = 50
classifications_to_replace = classification_counts[classification_counts < threshold].index
application_df['CLASSIFICATION'].replace(classifications_to_replace, 'Other', inplace=True)
```

```python
threshold = 50
classifications_to_replace = classification_counts[classification_counts < threshold].index
application_df['CLASSIFICATION'].replace(classifications_to_replace, 'Other', inplace=True)

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns = 'IS_SUCCESSFUL')
X = pd.get_dummies(X)
X
```

```python
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)


# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(keras.layers.Dense(55, activation='relu'))

# Second hidden layer
nn.add(keras.layers.Dense(55, activation='relu'))

# Output layer
nn.add(keras.layers.Dense(1, activation='sigmoid'))

nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

nn.fit(X_train_scaled, y_train, epochs=10)


# Check the structure of the model
nn.summary()
```
